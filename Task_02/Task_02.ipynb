# **Unemployment Analysis With Python**

Unemployment analysis in India using Python typically involves exploring and visualizing data related to employment and unemployment rates. The analysis can provide insights into the current state of unemployment in different regions, time trends, and potential factors influencing unemployment rates.

Data Preprocessing:
Clean and preprocess the data to ensure it is in a suitable format for analysis. This may involve handling missing values, data formatting, and converting data types.

Data Exploration:
Use libraries like Pandas and NumPy to explore the dataset and get a sense of its structure. Compute summary statistics, check for outliers, and understand the distribution of unemployment rates.

Data Visualization:
Utilize Python visualization libraries like Matplotlib, Seaborn, or Plotly to create various charts and graphs. Some common visualizations for unemployment analysis include line plots to visualize time trends, bar plots to compare unemployment rates across regions, and heatmaps to explore correlations between different economic indicators and unemployment.

Geospatial Analysis:
If the dataset includes geographical information, you can perform geospatial analysis to map the unemployment rates across different regions of India. Libraries like Geopandas and Plotly can be used for this purpose.

```python
import calendar
import datetime as dt
```

```python
un2['Date'] = pd.to_datetime(un2['Date'],dayfirst=True)
```

```python
un2['Month'] =  un2['Date'].dt.month
```

```python
un2['Month_int'] = un2['Month'].apply(lambda x : int(x))
```

```python
un2['Month_name'] =  un2['Month_int'].apply(lambda x: calendar.month_abbr[x])
```

```python
un2['Region'] = un2['Region'].astype('category')
```

```python
un2.drop(columns='Month',inplace=True)
```

```python
un2.head(1)
```

```python
sns.heatmap(un2.corr(),annot=True,cmap='coolwarm')
```

```python
sns.boxplot(x='Month_int',y='Estimated Unemployment Rate',data=un2)
```

```python
import plotly.express as px

fig = px.scatter_matrix(un2, template='plotly',
                        dimensions=['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate'],
                        color='Region')
fig.show()
```

```python
plot_ump =un2[['Estimated Unemployment Rate','States']]

df_unemp = plot_ump.groupby('States').mean().reset_index()

df_unemp = df_unemp.sort_values('Estimated Unemployment Rate')

fig = px.bar(df_unemp, x='States',y='Estimated Unemployment Rate',color='States',
            title='Average Unemployment Rate in each state',template='plotly')

fig.show()
```
